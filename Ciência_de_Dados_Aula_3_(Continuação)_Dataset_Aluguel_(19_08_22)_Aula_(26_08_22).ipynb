```python
#Filtrando os tipos para residencial
residencial = ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila']
```

```python
#Filtro para uma seleção de valores booleanos
selecao = dados['Tipo'].isin(residencial)
dados_residenciais = dados[selecao]
```

```python
#Comando para filtrar a base dados com os tipos requeridos
print('A base de dados original tem {} registros e a base nova tem {} registros'.format(dados.shape[0], dados_residenciais.shape[0]))
```

```python
#Resetando o índice e excluindo (dropando) o index criado indevidamente
dados_residenciais.reset_index(inplace=True)
dados_residenciais.pop('index')
dados_residenciais
```

```python
#Salvando a base de dados
dados_residenciais.to_csv('Aluguel_Residencial.csv', sep=';', index=False)
```

```python
#Carregando a base de dados salva anteriormente
dados_residenciais=pd.read_csv('Aluguel_Residencial.csv', sep=';')
```

```python
dados_residenciais
```

```python
#Verificando dados nulos na coluna 'Valor' e visualizando através da variável 'selecao'
selecao=dados_residenciais['Valor'].isnull()
dados_residenciais[selecao]
```

```python
#Quantidade de dados, levando em consideração os nulos
dados_residenciais.shape
```

```python
#Código para excluir dados nulos (dropna)
#Quantidade de dados, não levando em consideração os nulos
dados_residenciais.dropna(subset=['Valor'], inplace=True)
dados_residenciais.shape
```

```python
#Descobrindo dados nulos em 'Condominio'
selecao=dados_residenciais['Condominio'].isnull()
dados_residenciais[selecao]
```

```python
#Preenchendo valores nulos em Condomínio e IPTU com 0 (zeros)
dados_residenciais[selecao]=dados_residenciais[selecao].fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados_residenciais
```

```python
dados_residenciais.filter(items=[22558, 22579], axis=0)
```

```python
#Criando uma nova coluna na base de dados
dados_residenciais['Valor Comissão']=dados_residenciais['Valor']*0.10
dados_residenciais
```

```python
#Média do IPTU
dados_residenciais.IPTU.mean()
```

```python
#Filtrando
dados_residenciais.IPTU.fillna(dados_residenciais.IPTU.mean())
```

```python
dados_residenciais.IPTU.fillna(method='ffill', inplace=True)
dados_residenciais
```
