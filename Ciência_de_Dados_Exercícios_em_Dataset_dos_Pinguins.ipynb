```python
#pinguim2.drop(columns=['studyName'], inplace=True)
#pinguim2.drop(columns=['Region'], inplace=True)
#pinguim2.drop(columns=['Island'], inplace=True)
#pinguim2.drop(columns=['Stage'], inplace=True)
#pinguim2.drop(columns=['Individual ID'], inplace=True)
#pinguim2.drop(columns=['Clutch Completion'], inplace=True)
#pinguim2.drop(columns=['Date Egg'], inplace=True)
#pinguim2.drop(columns=['Sex'], inplace=True)
#pinguim2.drop(columns=['Comments'], inplace=True)
pinguim2.dropna(inplace=True)
```

```python
#Remover a classe dos atributos (no X)
X = pinguim2.drop(columns = ['Species']).values
#Pegar somente a classe (no Y)
y = pinguim2['Species'].values
```

```python
from sklearn.model_selection import train_test_split #Biblioteca para auxiliar na serapação de dados para treino e dados para testes
from sklearn.neighbors import KNeighborsClassifier #Biblioteca Python para trabalhos com Machine Learning
```

```python
#Separação dos dados
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train, Y_train) #Treinando algoritmo (Fit -> Treinar)
```

```python
result = model.predict(X_test)
result
```

```python
display(result)
display(Y_test)
```

```python
display(result==Y_test)
```

```python
total = result.shape[0]
total
```

```python
#Descobrindo a quantidade de dados verdadeiros
verdadeiros = 0
for i in result==Y_test:
  if i==True:
    verdadeiros =  verdadeiros + 1
print(verdadeiros)
```

```python
#Descobrindo a taxa de acerto (1 quer dizer 100%)
acuracia = verdadeiros/total
acuracia
```
