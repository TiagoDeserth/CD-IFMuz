```python
import pandas as pd
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
dados.head()
```

```python
#Linhas | Colunas
dados.shape
```

```python
dados.iloc[[], [dados.shape[1]-1]].columns.values.tolist()[0]
```

```python
#Data atualizando em tempo real
last_date = str(dados.iloc[[], [dados.shape[1]-1]].columns.values.tolist()[0])
last_date
```

```python
#Países contitdos no dataset
dados["Country/Region"].unique()
```

```python
dados.iloc[0]
```

```python
dados["Province/State"].unique()
```

```python
dados.set_index("Province/State").loc['Beijing']
```

```python
dados["Country/Region"].value_counts()
```

```python
confirmados_por_pais=dados.groupby("Country/Region").sum()
confirmados_por_pais
```

```python
confirmados_por_pais.loc['Brazil'][2:]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
confirmados_por_pais.loc['Brazil'][2:].plot()
plt.title("Casos confirmados no Brasil")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
confirmados_por_pais.loc['Brazil'][2:].plot()
plt.title("Casos confirmados no Brasil")
plt.ylim(0,4000000)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
confirmados_por_pais.loc['China'][2:].plot()
plt.title("Casos confirmados na China")
plt.show()
```

```python
confirmados_por_pais[last_date].sort_values(ascending=False)[:20].plot(kind='bar')
```

```python
#Diferença entre dias
confirmados_por_pais.loc['Brazil'][2:].diff()
```

```python
#Excluindo o nulo
confirmados_por_pais.loc['Brazil'][2:].diff().dropna()
```

```python
plt.figure(figsize=(10,5))
confirmados_por_pais_brazil = confirmados_por_pais.loc['Brazil'][2:].diff().dropna() 
confirmados_por_pais_brazil.plot()
```

```python
plt.figure(figsize=(10,5))
confirmados_por_pais_china = confirmados_por_pais.loc['China'][2:].diff().dropna() 
confirmados_por_pais_china.plot()
```

```python
#Exemplo de gráfico ruim
plt.figure(figsize=(10,5))
confirmados_por_pais[last_date].plot()
plt.title("Confirmados por país")
```

```python
#Exemplo de gráfico RUIM
plt.figure(figsize=(10,5))
confirmados_por_pais[last_date].plot(kind='pie')
plt.title("Confirmados por país")
plt.show()
```

```python
#Exemplo de gráfico RUIM
plt.figure(figsize=(10,5))
confirmados_por_pais[last_date].plot(kind='bar')
plt.title("Confirmados por país")
```

```python
#Exemplo de gráfico BOM
plt.figure(figsize=(10,5))
confirmados_por_pais[last_date].sort_values(ascending=False)[:10].plot(kind='bar')
plt.title("Confirmados por país")
plt.show()
```
