###Taxa de incidÃªncia 
Um evento incidente, no caso de uma nova doenÃ§a Ã© aquele ocorrido em um determinado tempo de observaÃ§Ã£o.
A taxa de incidÃªncia Ã© definida como o nÃºmero de casos novos divido pela populaÃ§Ã£o em risco da doenÃ§a em um espaÃ§o geogrÃ¡fico em comum.

```python
un_population = pd.read_csv("https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2022_TotalPopulationBySex.csv")
un_population['PopTotal'] = un_population['PopTotal']*1000
un_population.head()
```

```python
un_population.Location.unique()
```

```python
un_population_total = un_population.query("Time==2019")[["Location", "PopTotal"]].set_index("Location")
```

```python
un_population_total.head()
```

```python
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

confirmed = confirmed.iloc[:, :index_date+1]
deaths = deaths.iloc[:, :index_date+1]
recovered = recovered.iloc[:, :index_date+1]
```

```python
def lasted_by_country(data):
  last_date = data.iloc[[], [-1]].columns.values.tolist()[0]
  return data.groupby("Country/Region").sum()[last_date]
```

```python
informations = [lasted_by_country(confirmed), lasted_by_country(deaths), lasted_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ["confirmed", "deaths", "recovered"]
combined.head()
```

```python
for i in combined.index:
  print(i)
```

```python
un_population.query("Location.str.contains('Korea') and Time==2019", engine='python') #Contains - pesquisa por string
```

```python
def rename_location(location):
  #print(location)
  if location == "Republic of Korea":
    return 'Korea, South'
  if location == "Iran (Islamic Republic of)":
    return 'Iran'
  return location

un_population_total['location_for_who'] = un_population_total.index.map(rename_location)
un_population_total.head()      
```

```python
un_population_total.loc['Republic of Korea']
```

```python
un_population_total = un_population_total.set_index('location_for_who')
combined_expanded = combined.join(un_population_total).sort_values("confirmed", ascending = False).dropna()
```

```python
letality_rate1 = death_cases/confirmed_cases*100
letality_rate2 = death_cases/(death_cases+recovered_cases)*100

print(f"Lelity Rate 1: {letality_rate1}%")
print(f"Lelity Rate 2: {letality_rate2}%")
```

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:], recovered.sum()[3:], deaths.sum()[3:]], index = ['confirmed', 'recovered', 'deaths']).T
display(grouped_per_day.head())
display(grouped_per_day.tail())
```

```python
plt.figure(figsize=(10,5))
grouped_per_day["letality_rate_1"].plot() #Linha azul
grouped_per_day["letality_rate_2"].plot() #Linha laranja
plt.title("Letalidade pessimista e otimista no mundo todo")
plt.legend(['Mortes/Casos Confirmados', 'Mortes/(Mortes+Recuperados)'])
plt.grid(linestyle = '-')
plt.xlabel("Data")
plt.ylabel("% de Letalidade")
plt.show()
```
